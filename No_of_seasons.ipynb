```python
title_basics = pd.read_csv(r'/content/drive/My Drive//title.basics.tsv.gz', sep = '\t')
title_basics.head()
```

```python
title_basics.titleType.unique() #to look for tvseries
```

```python
tv = title_basics.titleType.unique()[3:8] #keeping tv types
tv
```

```python
tv_series = title_basics[title_basics.titleType.isin(tv)] #extracted tv types
tv_series
```

```python
tv_series_rn = tv_series[['tconst','startYear','endYear']] #keeping the reqd rows
tv_series_rn
```

```python
tv_series_rn = tv_series_rn.replace('\\N', np.nan) ## replacing \N
tv_series_rn = tv_series_rn.dropna() # dropping NaN
tv_series_rn = tv_series_rn.reset_index() 
tv_series_rn = tv_series_rn.drop('index', axis =1)
tv_series_rn
```

```python
tv_series_rn.dtypes
```

```python
tv_series_rn.startYear = tv_series_rn.startYear.astype('int32')
tv_series_rn.endYear = tv_series_rn.endYear.astype('int32')
```

```python
tv_series_rn['Years'] = tv_series_rn.endYear - tv_series_rn.startYear
tv_series_rn
```

```python
tv_series_rn = tv_series_rn[['tconst','Years']]
tv_series_rn
```

```python
ratings_title = pd.read_csv(r'https://datasets.imdbws.com/title.ratings.tsv.gz', sep = '\t') #to get a relation of nconst and tconst
ratings_title.head()
```

```python
#keeping only those having number of votes greater than 50
ratings_title = ratings_title[ratings_title.numVotes>=50]
ratings_title
```

```python
ratings_title = ratings_title.drop('numVotes', axis = 1)
ratings_title
```

```python
tv_dur_rat = tv_series_rn.merge(ratings_title, how = 'inner', on = 'tconst')
tv_dur_rat
```

```python
#assuming good serials to be greater than average rating 8 and just keeping them
tv_dur_rat_top = tv_dur_rat[tv_dur_rat.averageRating>=8]
tv_dur_rat_top = tv_dur_rat_top.reset_index()
tv_dur_rat_top = tv_dur_rat_top.drop('index', axis=1)
tv_dur_rat_top
```

```python
#keeping atlest greater than one year
tv_dur_rat_top = tv_dur_rat_top[tv_dur_rat_top.Years>=1]
tv_dur_rat_top = tv_dur_rat_top.reset_index()
tv_dur_rat_top = tv_dur_rat_top.drop('index', axis=1)
tv_dur_rat_top
```

```python
tv_dur_rat_top.sort_values(by = 'averageRating', ascending= False, inplace = True)
tv_dur_rat_top = tv_dur_rat_top.reset_index()
tv_dur_rat_top = tv_dur_rat_top.drop('index', axis=1)
tv_dur_rat_top.head(50)
```
